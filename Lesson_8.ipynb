```python
import csv
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
data_path = 'opendata.csv'
data = pd.read_csv(data_path, 
                       sep=',', encoding='cp1251')
```

```python
names = set(data['name'])
regions = set(data['region'])
name_input = input(f'Выберите одно из этих значений {names}:')
region_input = input(f'Выберите область {regions}:')
interval_from = input('Введите числo от, формат 2018-01-23:')
interval_to = input('Введите числo до, формат 2019-03-12:')
```

```python
indexes = data.loc[(data['name'] == name_input)&(data['region'] == region_input) & ((data['date'] > interval_from) & (data['date'] < interval_to))].index
plt.plot(data.loc[indexes]['date'], data.loc[indexes]['value']);
plt.title(name_input)
plt.xlabel('От ' + interval_from +'    '+ 'До ' + interval_to)
plt.xticks([]);
```
